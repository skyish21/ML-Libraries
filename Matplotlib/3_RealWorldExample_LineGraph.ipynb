```python
gas= pd.read_csv(r"D:\python\Matplotlib\gas_prices.csv")

plt.figure(figsize=(5,3))

plt.title('Yearly Gas Prices for different Countries')

for country in gas:
    if country != 'Year':
       #add plot properties for all
       plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xlabel('Year') 
plt.ylabel('Gas Prices (USD)')

plt.xticks(gas.Year[::3])

plt.legend()

plt.show()
```

```python
gas= pd.read_csv(r"D:\python\Matplotlib\gas_prices.csv")

plt.figure(figsize=(5,3))

plt.title('Yearly Gas Prices for different Countries')

for country in gas:
    if country != 'Year':
       #add plot properties for all
       plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xlabel('Year') 
plt.ylabel('Gas Prices (USD)')

plt.xticks(gas.Year[::3])

# Move the legend outside the plot
#bbox_to_anchor=(1.05, 1): Moves the legend box slightly to the right of the plot. Adjust 1.05 as needed to control how far outside the plot it goes.
#loc='upper left': Anchors the legend to the upper-left corner of the bounding box (as set by bbox_to_anchor).
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
gas= pd.read_csv(r"D:\python\Matplotlib\gas_prices.csv")

plt.figure(figsize=(5,3))

#change font
plt.title('Yearly Gas Prices for different Countries', fontdict={'fontweight':'bold','fontsize':15})

#add selected countries
pick_countries=['Australia','Canada','France']
for country in gas:
    if country in pick_countries:
       plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xlabel('Year') 
plt.ylabel('Gas Prices (USD)')

#gas.Year[::3]+[2011] == direct list cannot be added
#gas.Year[::3] is a dataframe so convert to list to add new value
plt.xticks(gas.Year[::3].tolist()+[2011])

plt.legend()

#save figure with high res(dpi=pixels per inch)
plt.savefig('Gas_Price_Graph.png' ,dpi=300)

plt.show()
```
