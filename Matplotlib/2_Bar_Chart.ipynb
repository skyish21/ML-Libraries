```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Bar Chart


```python
labels=['A','B','C']
values=[2,4,6]

plt.bar(labels,values)

plt.show()
```

```python
labels=['A','B','C']
values=[2,4,6]

bars=plt.bar(labels,values)
#Hatching (pattern filled polygons) can be added to most polygons in Matplotlib, including bar, fill_between, contourf, and children of Polygon
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.show()
```

```python
labels=['A','B','C']
values=[2,4,6]

bars=plt.bar(labels,values)
#alternate way to design bar
patterns=['/','o','*']

for bar in bars:
    #pop the last element first (into the first bar)
    bar.set_hatch(patterns.pop())
    
plt.show()
```

```python
labels=['A','B','C']
values=[2,4,6]

bars=plt.bar(labels,values)
patterns=['/','o','*']

for bar in bars:
    #pop the 0 index(first element) first (into the first bar)
    bar.set_hatch(patterns.pop(0))
    
plt.show()
```
