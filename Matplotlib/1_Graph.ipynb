```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
#Plot y versus x as lines and/or markers.
#plot x and y using default line style and color
x=[0,1,2,3,4]
y=[0,2,4,6,8]
plt.plot(x,y)

#Set a title for the Axes.
#specify font parameters with fontdict
plt.title("First Graph", fontdict={'fontname': 'Comic Sans MS', 'fontsize':20})

#Set the label for the x-axis and y-axis.
plt.xlabel('X Axis', fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('Y Axis')

#show plot
plt.show() 
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
plt.plot(x,y)

#x and y tickmarks. Scale of the graph
#remove decimal values
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#key augmentation notation
#plot properties
#color or c and can pass hexdecimal color values
plt.plot(x,y, label='y=2x', color='pink', linewidth=2, linestyle='--', marker='.', markersize='10', markeredgecolor='red')

#place a legend on the axes.
#legend on a graph or plot is a small box that explains what each color, line, or symbol represents
#displays the legend with the label "y=2x" on the plot.
plt.legend()

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#shorthand notation to use plot properties to customize graph
#fmt='[color][marker][line]'
plt.plot(x,y, 'r.--', label='2x')

plt.legend()

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#line number 1
plt.plot(x,y, 'r.--', label='2x')

#line number 2

#select interval we want to plot points at
#The np.arange() function generates a sequence of numbers with a specified start, stop, and step(tick)
#(start,stop,step(tick value))
l2=np.arange(0,4.5,0.5)

#l2 = x-axis values and l2**2 = y-axis values
plt.plot(l2,l2**2, 'b', label='l^2')

plt.legend()

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y, 'r.--', label='2x')

#line number 2
l2=np.arange(0,4.5,0.5)

#plot part of graph before six as line
plt.plot(l2[:6], l2[:6]**2,'b', label='l^2')

#plot remainder of graph from 5 as line
plt.plot(l2[5:], l2[5:]**2,'b--')

plt.legend()

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#resize your image
#(dpi=pixels per inch)>=300 for high resolution image and less pixalation
plt.figure(figsize=(5,3), dpi=100)
plt.plot(x,y, 'r.--', label='2x')

#line number 2
l2=np.arange(0,4.5,0.5)
plt.plot(l2[:6], l2[:6]**2,'b', label='l^2')
plt.plot(l2[5:], l2[5:]**2,'b--')

plt.legend()

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

#save your graph with high res with dpi 300
plt.savefig('mygraph.png', dpi=300)

plt.show()
```
